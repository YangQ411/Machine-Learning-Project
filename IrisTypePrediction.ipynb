## **1. Import libraries**

```python
import numpy as np # for calculating 
import pandas as pd # for visualization and contruction of dadaframe
import matplotlib as plt # for plotting
```

## **2. loading dataset**

```python
dataset = pd.read_csv('iris.csv')
dataset
```

### The types of output

```python
dataset.Species.value_counts(normalize=False)
```

## **3. Split dataset into train and test**

```python
from sklearn.model_selection import train_test_split

X = dataset.drop(['Species'],axis=1)
Y = dataset['Species']  

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, 
                                                    test_size=0.4, 
                                                    random_state=1) 
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

## **5. Train our dataset**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
Clf = RandomForestClassifier(n_jobs = 2, random_state = 0)
Clf.fit(X_train, Y_train)
```

## **6. Make Prediction**

```python
y_pred = Clf.predict(X_test)
```

## **7. Evaluate our Model**

```python
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score

print('Accuracy score of Random Froest: ', accuracy_score(Y_test, y_pred)*100)
print('f1 score of Random Froest: ', f1_score(Y_test, y_pred, average='weighted'))
```
